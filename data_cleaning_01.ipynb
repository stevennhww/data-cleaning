## Clean a txt file through a function

```python
university_towns = []
with open ('/Users/huimeixiao/Downloads/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            university_towns.append((state,line))
```

```python
university_towns[:10]
```

```python
towns_df = pd.DataFrame(university_towns, columns = ['state','region'])
towns_df.head()
```

## Using applymap to clean the df with a function

```python
def find_citystate(item):
    if '(' in item:
        return item[:item.find('(')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(find_citystate)
towns_df.head()
```
